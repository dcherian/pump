````python
def generate_mixpods_diag_table(casename, section_name, section):
    prefix = f"{casename}.mom6"

    # extents for constant lon
    # lat_min, lat_max = -6, 6  # 100, 500

    # TODO: generalize
    # extents for constant lat
    # lon_min, lon_max = 240 - 0.15, 283 + 0.2  # 100, 800

    extentstr = f"{section.extent[0]} {section.extent[1]}"

    top = dict()

    delta = 0

    line = f"# {section_name} section for miχpods\n"
    for index_name, locations in section.index.items():
        index = geometry.indexes[index_name]
        for loc in locations:
            (i,) = index.get_indexer([loc], method="nearest")
            if index_name == "lonh":
                regionstr = f'"{index[i-1]:.3f} {index[i+1]:.3f} {extentstr} -1 -1"'
                region_in_file = f"{-1 * loc:.0f}W"
                # for idx in index[np.arange(i-delta, i+delta+1)]:
                #    plt.plot([idx, idx], section.extent, lw=0.5, color='k')

            elif index_name == "lath":
                # TODO: This doesn't work , I don't think
                regionstr = (
                    f'"{extentstr} {index[i-delta]:.3f} {index[i+delta]:.3f} -1 -1"'
                )
                for idx in index[np.arange(i - delta, i + delta + 1)]:
                    plt.plot(section.extent, [idx, idx], lw=0.5, color="k")

            fname = f"{prefix}.{section_name}_{region_in_file.replace(' ', '_')}_%4yr"
            if fname not in top:
                top[fname] = f'"{fname}", 1, "hours", 1, "hours", "time", 1, "years"'
            for var in MIXPODS_VARS:
                line += f'"ocean_model", "{var}", "{var}", "{fname}", "all", "mean", {regionstr}, 2\n'
                # print(fname)

    return top, line


toptao, tao = generate_mixpods_diag_table(
    casename,
    section_name="TAO",
    section=Section(
        index={"lonh": np.array([-170, -155, -140, -125, -110, -90])}, extent=(-6, 6)
    ),
)

toppirata, pirata = generate_mixpods_diag_table(
    casename,
    section_name="PIRATA",
    section=Section(index={"lonh": np.array([-23, -10])}, extent=(-6, 6)),
)

toprama80, rama80 = generate_mixpods_diag_table(
    casename,
    section_name="RAMA_80E",
    section=Section(index={"lonh": np.array([80.5 - 360])}, extent=(-3, 6)),
)

toprama90, rama90 = generate_mixpods_diag_table(
    casename,
    section_name="RAMA_90E",
    section=Section(index={"lonh": np.array([90 - 360])}, extent=(-6, 23)),
)
````

````python
%matplotlib inline

#prefix = "gmom.e23.GJRAv4.TL319_t061_hycom1_N75.nuopc.baseline.002"

section_name = "TAO"

varnames = [
    # state
    "uo", "vo", "zos",
    "thetao", "so",
    "taux", "tauy",
    "net_heat_surface", 
    "SW", 
    "N2",
    "Tflx_dia_diff", #"Sflx_dia_diff",
    # Total
    "Kd_heat", # or 
    "Kv_u", "Kv_v",
    "mlotst", # mixed layer
    # KPP
    "KPP_kheat", # needed?
    "KPP_OBLdepth",
    "KPP_buoyFlux",
    "KPP_N2",
    "KPP_NLtransport_heat",
    "N2_shear",
    "S2_shear",
    # ePBL
    "Kd_ePBL",
    "ePBL_h_ML",
    # "e"? interface depths; hm files
    "volcello",  # required with h
    "h",
    "ri_grad_shear_orig", "ri_grad_shear",
] + wci_vars

sections = {
    "lonh": np.array([-170, -155, -140, -125, -110, -90]),
    #"lath": [],
}

# extents for constant lon
lat_min, lat_max = -6, 6    # 100, 500
# extents for constant lat
lon_min, lon_max = 240-0.15, 283+0.2 # 100, 800

top = dict()

delta = 0

line = "# eq Pac TAO sections\n"
for index_name, locations in sections.items():
    index = geometry.indexes[index_name]
    for loc in locations:
        i, = index.get_indexer([loc], method="nearest")
        if index_name == "lonh":
            regionstr = f'"{index[i-1]:.3f} {index[i+1]:.3f} {lat_min} {lat_max} -1 -1"'
            region_in_file = f"{-1 * loc:.0f}W"
            #for idx in index[np.arange(i-delta, i+delta+1)]:
            #    plt.plot([idx, idx], [lat_min, lat_max], lw=0.5, color='k')

        elif index_name == "lath":
            regionstr = f'"{lon_min} {lon_max} {idx[i-delta]:.3f} {index[i+delta]:.3f} -1 -1"'
            for idx in index[np.arange(i-delta, i+delta+1)]:
                plt.plot([lon_min, lon_max], [idx, idx], lw=0.5, color='k')

        fname = f"{prefix}.{section_name}_{region_in_file.replace(' ', '_')}_%4yr"
        if fname not in top:
            top[fname] = f'"{fname}", 1, "hours", 1, "hours", "time", 1, "years"'
        for var in varnames:
            line += f'"ocean_model", "{var}", "{var}", "{fname}", "all", "mean", {regionstr}, 2\n'
            # print(fname)

print(header + "\n".join(top.values()))
print("\n\n")
print(section2 + "\n\n" + line)
print("\n")
with open(f"/glade/work/dcherian/cesm/{casename}/SourceMods/src.mom/diag_table", "w") as file:
    file.write(header + "\n".join(top.values()) + section2 + "\n\n" + line)
````
